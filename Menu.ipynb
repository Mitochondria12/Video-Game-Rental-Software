```python
#James Sharpe

from database import *
import ipywidgets as widgets
from IPython.display import display, clear_output
import matplotlib.pyplot as plt
#This creates the buttons for our user interface.
button1 = widgets.Button(description="Search for game and its avaliability.")
button2 = widgets.Button(description="Rent a Game to a Customer.")
button3 = widgets.Button(description="Return a Game.")
button4 = widgets.Button(description="Find out the best games to buy for a certain month.")
button5=widgets.Button(description="See most popular game rentals for a certain month")

# This creates a container for the entire UI everything appears here.
ui_container = widgets.VBox([]) 


# This is a function to display the main menu.
def display_main_menu(b=None):
    ui_container.children = [button1, button2, button3, button4,button5]

#This is a button which allows us to return to the main menu.
back_to_menu_button = widgets.Button(description="Back to Main Menu")
back_to_menu_button.on_click(display_main_menu)

#This is Button 1. When it gets clicked on it opens up a
game_title_input = widgets.Text(description="Game Title:")
game_platform_input = widgets.Dropdown(
    options=['PC', 'Xbox', 'Nintendo Switch', 'PlayStation'],
    description='Platform:',
    value='PC',  
    disabled=False,  
)
submit_button = widgets.Button(description="Search for game data.")
search_output = widgets.Output()

def on_submit_clicked(b):
    with search_output:
        search_output.clear_output()
        game_title = game_title_input.value
        game_platform = game_platform_input.value
        info = get_available_games_info(game_title, game_platform)
        print(info)

submit_button.on_click(on_submit_clicked)

game_info_form = widgets.VBox([game_title_input, game_platform_input, submit_button, search_output, back_to_menu_button])

def on_button1_clicked(b):
    ui_container.children = [game_info_form]

#Button 2
customer_id_input = widgets.Text(description="Customer Id:")
game_id_input=widgets.Text(description="Game Id:")
rent_submit_button = widgets.Button(description="Submit Rental Request")
rent_form_output = widgets.Output()

def rent_on_submit_clicked(b):
    with rent_form_output:
        rent_form_output.clear_output()
        customer_id = (customer_id_input.value)
        game_id = (game_id_input.value)
        message = can_customer_rent_another_game(customer_id, game_id)
        print(message)

rent_submit_button.on_click(rent_on_submit_clicked)

rental_info_form = widgets.VBox([customer_id_input, game_id_input, rent_submit_button, rent_form_output, back_to_menu_button])

def on_button2_clicked(b):
    ui_container.children = [rental_info_form]
    
#Button 3
return_game_id_input = widgets.Text(description="Game Id:")
return_submit_button = widgets.Button(description="Submit Game Search")
return_form_output = widgets.Output()

def return_on_submit_clicked(b):
    with return_form_output:
        return_form_output.clear_output()
        game_id = return_game_id_input.value
        message = returning_game(game_id)
        print(message)

return_submit_button.on_click(return_on_submit_clicked)

return_form = widgets.VBox([return_game_id_input, return_submit_button, return_form_output, back_to_menu_button])


def on_button3_clicked(b):
    ui_container.children = [return_form]


#Button 4
best_games_budget_input = widgets.Text(description="Budget:")
best_games_month_input = widgets.Text(description="Month:")
best_games_submit_button = widgets.Button(description="Submit best games to buy query.")
best_games_form_output = widgets.Output()

def best_games_on_submit_clicked(b):
    with best_games_form_output:
        best_games_form_output.clear_output()
        budget = best_games_budget_input.value
        month=best_games_month_input.value
        message = games_to_buy(budget,month)
        print(message)

best_games_submit_button.on_click(best_games_on_submit_clicked)

best_games = widgets.VBox([best_games_budget_input, best_games_month_input,best_games_submit_button, best_games_form_output, back_to_menu_button])


def on_button4_clicked(b):
    ui_container.children = [best_games]

#Button 5
title_month_input = widgets.Text(description="Month:")
title_submit_button = widgets.Button(description="Submit Best Games Month Search")
title_form_output = widgets.Output()

def graph_on_submit_clicked(b):
    with title_form_output:
        title_form_output.clear_output()
        title_month = title_month_input.value
        graph = most_popular_title_for_month_graph(title_month)
        plt.show()

title_submit_button.on_click(graph_on_submit_clicked)

title_form = widgets.VBox([title_month_input, title_submit_button, title_form_output, back_to_menu_button])


def on_button5_clicked(b):
    ui_container.children = [title_form]


# Link buttons to functions
button1.on_click(on_button1_clicked)
button2.on_click(on_button2_clicked)
button3.on_click(on_button3_clicked)
button4.on_click(on_button4_clicked)
button5.on_click(on_button5_clicked)
# Display UI
display(ui_container)

# Initial display of buttons
display_main_menu()
```
